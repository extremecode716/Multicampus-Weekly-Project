```python
# Sample 2
# ÌïúÍµ≠ ÏõπÌà∞, Ï†ú2Ïùò Ìï¥Î¶¨Ìè¨ÌÑ∞Í∞Ä Îê† Ïàò ÏûàÏùÑÍπå?
# https://www.youtube.com/watch?v=5z1N5TRNchI&ab_channel=%EC%8A%88%EC%B9%B4%EC%9B%94%EB%93%9C

video_id = '5z1N5TRNchI'
title = 'ÌïúÍµ≠ ÏõπÌà∞, Ï†ú2Ïùò Ìï¥Î¶¨Ìè¨ÌÑ∞Í∞Ä Îê† Ïàò ÏûàÏùÑÍπå'
comments = []
results = youtube.commentThreads().list(
    videoId = video_id,
    order='time',
    part = 'snippet',
    textFormat='plainText',
    maxResults = 100
    ).execute()
```

```python
while results:
    for item in results['items']:
        comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
        comments.append(comment)
        
    if 'nextPageToken' in results:
            pageToken = results['nextPageToken']
            results = youtube.commentThreads().list(
                videoId = 'X7-aZpsnw5M',
                order='time',
                part = 'snippet',
                textFormat='plainText',
                pageToken = pageToken,
                maxResults = 100
                ).execute()
    else:
        break
```

```python
comments_str = ' '.join(comments)
print(comments_str)
```

```python
# ave as string
DATA_PATH = './data/'
if not os.path.exists(DATA_PATH):
    os.makedirs(DATA_PATH)
filepath = '{title}_{video_id}.txt'.format(title=title,video_id=video_id)
f = open(DATA_PATH + filepath, 'w')
f.write(comments_str)
f.close()

print(filepath + " saved.")
```
