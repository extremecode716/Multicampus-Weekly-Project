```python
service = Service(executable_path=ChromeDriverManager().install())
```

```python

def get_comments(service, df):
    TIMEOUT = 1   
    driver = webdriver.Chrome(service=service)
    
    url = 'http://www.yes24.com/Product/Goods/'
    episode_url = url + str(yes_url_01)
    print(episode_url)
    driver.get(episode_url) # 새창 열기
    driver.implicitly_wait(TIMEOUT)
       

#     전체 댓글 클릭
    driver.find_element(By.CSS_SELECTOR,'#total > a > span').click()
    time.sleep(1)
    
    # "맨뒤" 클릭하여 전체 페이지 수 조회
    driver.find_element(By.CSS_SELECTOR,'#infoset_reviewContentList > div.review_sort.sortTop > div.review_sortLft > div > a.bgYUI.end').click()
    time.sleep(1)
    strong = driver.find_element(By.CSS_SELECTOR,'#infoset_reviewContentList > div.review_sort.sortTop > div.review_sortLft > div > strong')
    num_pages =strong.text  # class_ = a.get_attribute('class')
    print("num_pages: ", num_pages)     
    
    # "맨앞" 으로 이동
    driver.find_element(By.CSS_SELECTOR,'#infoset_reviewContentList > div.review_sort.sortTop > div.review_sortLft > div > a').click()
    time.sleep(5)
    
    #댓글 수집
    for page in range(int(num_pages)):

        #최근 열린 탭으로 전환
        driver.switch_to.window(driver.window_handles[-1]) 
        soup = bs(driver.page_source,"html.parser")

      

       ############## # 더보기 클릭################################
        try:
            review_click1=driver.find_element(By.CSS_SELECTOR,'#infoset_reviewContentList > div:nth-child(3) > div.reviewInfoBot.crop > a > div > span').click()
            time.sleep(1)
            review_click2=driver.find_element(By.CSS_SELECTOR,'#infoset_reviewContentList > div:nth-child(4) > div.reviewInfoBot.crop > a > div > span').click()
            time.sleep(1)
            review_click3=driver.find_element(By.CSS_SELECTOR,'#infoset_reviewContentList > div:nth-child(5) > div.reviewInfoBot.crop > a > div > span').click()
            time.sleep(1)
            review_click4=driver.find_element(By.CSS_SELECTOR,'#infoset_reviewContentList > div:nth-child(6) > div.reviewInfoBot.crop > a > div > span').click()
            time.sleep(1)
            review_click5=driver.find_element(By.CSS_SELECTOR,'#infoset_reviewContentList > div:nth-child(7) > div.reviewInfoBot.crop > a > div > span').click()
           
        except: 
              pass
            
            
        reviews = soup.findAll("div", {"class":["review_cont"]})
        reviews = [review.get_text().replace('\n','') for review in reviews]
        df_page = pd.DataFrame(data=[], columns=['Review'])
        df_page['Review'] = reviews
         

        
        df = pd.concat([df,df_page])      
        print(page+1, " page Done.")   
        display(df)

        if (page+1) == int(num_pages):
            return df
        
        # 다음 페이지로 이동
        a=driver.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div.review_sort.sortTop > div.review_sortLft > div > strong+a')
        a.send_keys('\n')
        time.sleep(10)
    driver.close()
    return df

df = pd.DataFrame(data=[], columns=['Review'])
df = get_comments(service, df) 
df

```
