```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup as bs
import pandas as pd
import os
import time
import re
```

```python
pd.set_option("display.max_rows", 1000)
pd.set_option("display.max_columns", 1000)
pd.options.display.float_format = '{:.2f}'.format
```

```python
TIMEOUT = 1   
service = Service(executable_path='c:\Temp\chromedriver.exe')
driver = webdriver.Chrome(service=service)

# 나혼자만 레벨업
episode_url = 'https://webtoon.kakao.com/content/%EB%82%98-%ED%98%BC%EC%9E%90%EB%A7%8C-%EB%A0%88%EB%B2%A8%EC%97%85/2320'
print(episode_url)
driver.get(episode_url)
driver.implicitly_wait(TIMEOUT)
```

```python
# 세부 정보로 이동
driver.find_element(By.XPATH, '//*[@id="root"]/main/div/div[1]/div/div[1]/div[2]/div[1]/div[1]/div[3]/div[2]/div/div').click()
time.sleep(1)

# 댓글로 이동
driver.find_element(By.XPATH, '//*[@id="root"]/main/div/div[1]/div/div[1]/div[3]/div/div/div[1]/div/div[1]/div[1]/div/div[2]/ul/li[3]/p').click()
time.sleep(1)

# 좋아요순 -> 최신순 
driver.find_element(By.XPATH, '//*[@id="root"]/main/div/div[1]/div/div[1]/div[3]/div/div/div[1]/div/div[3]/button').click()
time.sleep(1)
```
