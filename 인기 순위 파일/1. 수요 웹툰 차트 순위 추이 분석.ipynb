## 수요일 차트인 횟수

```python
df1=df[['날짜','순위','작품명']]
df1.head()
```

```python
df2=df1.copy()
df2.columns=['날짜','차트인 횟수','작품명']
df2.head()
```

```python
pdf1 = pd.pivot_table(df2,                # 피벗할 데이터프레임
                     index = '작품명',    # 행 위치에 들어갈 열
#                      columns = '순위',    # 열 위치에 들어갈 열
                     values = '차트인 횟수',     # 데이터로 사용할 열
                     aggfunc = 'count')   # 데이터 집계함수
pdf1=pdf1.sort_values(('차트인 횟수'), ascending = False)
pdf1
```

```python
# sns.set(font_scale=1.5)
sns.set(font="Malgun Gothic", 
        rc={"axes.unicode_minus":False},
        style='darkgrid',
        font_scale=1.5)
sns.heatmap(pdf1
            ,cmap='YlGn'
            ,annot=True
            ,fmt='d'
            ,linewidths=.5
            ,annot_kws={"size": 20})

# plt.savefig('차트인.png')
```

```python
## 3개 데이터 분리
a_df = df3.query('작품명=="나쁜사람"') 
a_visits = a_df['순위']
days = a_df['날짜']
 
b_df = df3.query('작품명=="내 남편과 결혼해줘"')
b_visits = b_df['순위']
 
c_df = df3.query('작품명=="일렉시드"')
c_visits = c_df['순위']

d_df = df3.query('작품명=="튜토리얼 탑의 고인물"')
d_visits = d_df['순위']

e_df = df3.query('작품명=="전지적 독자 시점"')
e_visits = e_df['순위']



 
fig = plt.figure(figsize=(14,8)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 설정
ax = fig.add_subplot() ## 그림 뼈대(프레임) 생성
 
ax.plot(days,a_visits,marker='>',linestyle='dashed',label='나쁜사람',color='green') ## 선그래프 생성
ax.plot(days,b_visits,marker='>',linestyle='dashed',label='내 남편과 결혼해줘') 
ax.plot(days,c_visits,marker='>',linestyle='dashed',label='일렉시드') 
ax.plot(days,d_visits,marker='>',linestyle='dashed',label='튜토리얼 탑의 고인물',color='purple') 
ax.plot(days,e_visits,marker='>',markersize=20,label='전지적 독자 시점',linewidth=5,color='red') 

plt.legend(loc='lower right') 
ax.legend() ## 범례
 
# plt.title('수요 웹툰 차트 순위 추이',fontsize=40) ## 타이틀 설정
plt.show()
```
